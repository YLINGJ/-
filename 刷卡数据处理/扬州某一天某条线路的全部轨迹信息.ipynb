```python
##合并不同文件夹下的同名文件
import os
def get(name, info):
    # 写入新文件夹
    with open(r"D:\项目\扬州\BUSGPS\车辆编号带线路编号\{}".format(name), "a",encoding='utf-8') as f:###有中文，注意指定编码形式
        f.write("\n".join(info))
# 原文件夹
path = r"D:\项目\扬州\BUSGPS\cheid_xianlu_id_13"
lists = os.listdir(path)
for l in lists:
    # 子文件夹路径
    p = os.path.join(path, l)
    ps = os.listdir(p)
    for p1 in ps:
        # 每个文件路径
        p2 = os.path.join(p, p1)
        print(os.path.getsize(p2))
        if os.path.getsize(p2)>20:
            # 读取每个文件内容
            with open(p2, "r",encoding='utf-8') as f:###有中文，注意指定编码形式
                lines = f.readlines()
                get(p1, lines)
```
