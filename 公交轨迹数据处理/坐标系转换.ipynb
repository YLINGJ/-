```python
#运行这个代码可以让Python不显示warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import geopandas
import geopandas as gpd
import pyproj
```

```python
#读取保存结果
BUS_GPS = pd.read_csv(r'F:\data\22_5_1\BUS_GPS.csv',encoding='gbk')#这样读取默认第一行是表头
#线路数据
line = pd.read_csv(r'F:\data\22_5_1\22路线路.csv',encoding='gbk')#默认第一行是表头
#站点数据
stop = pd.read_csv(r'F:\data\22_5_1\stations.csv',encoding='utf-8')#默认第一行是表头
```

```python
#转换坐标！！！！！该替换的要替换干净，又是浪费时间的一个下午
##设置转换的原坐标系
BUS_GPS.crs={'init': 'epsg:4326'}
stop.crs={'init': 'epsg:4326'}
line.crs={'init': 'epsg:4326'}
##指定投影坐标系
p1 = pyproj.Proj(init='epsg:2416') # 一个投影坐标系统EPSG Code
lons0=BUS_GPS.LONGITUDE.values
lats0 =BUS_GPS.LATITUDE.values
#transformer.transform(lat,lon)
buslons,buslats = p1(lons0,lats0)
lons=stop.LONGITUDE.values
lats =stop.LATITUDE.values
#transformer.transform(lat,lon)
stoplons,stoplats = p1(lons,lats)
lons2=line.LONGITUDE.values
lats2 =line.LATITUDE.values
#transformer.transform(lat,lon)
linelons,linelats= p1(lons2,lats2)
```

```python
##将数组转换为PD
BUS_GPS['lons']=pd.DataFrame(buslons)
BUS_GPS['lats']=pd.DataFrame(buslats)
stop['lons']=pd.DataFrame(stoplons)
stop['lats']=pd.DataFrame(stoplats)
line['lons']=pd.DataFrame(linelons)
line['lats']=pd.DataFrame(linelats)
```

```python
#生成线路的线要素
from shapely.geometry import LineString
dianlon=line['lons']
dianlat=line['lats']
dianlist=[]
for i in range(0,860):
    g=(dianlon[i],dianlat[i])
    dianlist.append(g)
    pass
dianlist
linestr = LineString(dianlist)
line['geometry']=linestr
line
```

```python
##生成站点的的点要素
from shapely.geometry import Point
lon=stop['lons']
lat=stop['lats']
stoplist=[]
for i in range(0,93):
    g=Point(lon[i],lat[i])
    stoplist.append(g)
    pass
stoplist = gpd.GeoDataFrame(stoplist)###转换为pd形式
list
stop['geometry']=stoplist
stop
```
